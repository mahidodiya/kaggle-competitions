```python
import numpy as np
import pandas as pd
```

```python
#load 
train= pd.read_csv('/kaggle/input/competitions/house-prices-advanced-regression-techniques/train.csv')
test= pd.read_csv('/kaggle/input/competitions/house-prices-advanced-regression-techniques/test.csv')

print(train.shape)
print(test.shape)

train.head(2)
```

```python
#column with highest missing value
train.isnull().sum().sort_values(ascending=False).head()
```

```python
cols_to_drop = ['PoolQC', 'MiscFeature', 'Alley', 'Fence']

train = train.drop(cols_to_drop, axis=1)
test = test.drop(cols_to_drop, axis=1)

X = train.drop('SalePrice', axis=1)
y = train['SalePrice']
```

```python
num_cols = X.select_dtypes(include=['int64', 'float64']).columns
cat_cols = X.select_dtypes(include=['object']).columns
```

```python
print('SalePrice' in num_cols)
```

```python
#Fill missing value in numerical column
X[num_cols] = X[num_cols].fillna(
    X[num_cols].median()
)

test[num_cols] = test[num_cols].fillna(
    X[num_cols].median()
)
```

```python
#Fill missing value in categorial column
X[cat_cols] = X[cat_cols].fillna('Missing')
test[cat_cols] = test[cat_cols].fillna('Missing')
```

```python
#Verify missing value
print(X.isnull().sum().sum())
print(test.isnull().sum().sum())
```

```python
#One-Hot Encode
X = pd.get_dummies(X)
test_encoded = pd.get_dummies(test)

X, test_encoded = X.align(
    test_encoded,
    join='left',
    axis=1,
    fill_value=0
)

print(X.shape)
print(test_encoded.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

model.fit(X,y)
```

```python
prediction = model.predict(test_encoded)
print(prediction)
```

```python
#create submission file
submission=pd.DataFrame({
    'Id':test['Id'],
    'SalePrice': prediction
})

submission.to_csv('submission_v2.csv',index=False)

submission.head()
```

```python
feature_importance = pd.DataFrame({
    'Feature':X.columns,
    'Importance':model.feature_importances_
})

feature_importance.sort_values(
    by='Importance',
    ascending=False
).head(15)
```
